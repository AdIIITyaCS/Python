## Advance operation 

```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
sales_data=np.array([
  [1,15000,18000,22000,25000],
  [2,12000,14000,16000,19000],
  [3,20000,23000,26000,30000],
  [4,18000,21000,24000,27000],
  [5,18500,16000,20500,30000]
])
```

```python
print("========   ZOMATO SALES ANALYSIS   ============\n")
print("1st three company sales:   \n",sales_data[:3])
print("All company without their IDs :   \n",sales_data[:,1:])
```

```python
# total sales per year
print("Each year each company sell\n",np.sum(sales_data[:,1:],axis=0)) # to add data column wise(axis=0)
print("Whole 4 years Revenue of each company\n",np.sum(sales_data[:,1:],axis=1)) # to add data row wise(axis=1)

```

```python
# minimum sales per company
min_sales=np.min(sales_data[:,1:],axis=1)
print("Minimum Sales per company\n",min_sales)

# maximum sales per year
min_sales=np.max(sales_data[:,1:],axis=0)
print("Maximum Sales per year\n",min_sales)

# average
average_sales=np.mean(sales_data[:,1:],axis=1)
print("Average Sales per company\n",average_sales)

# cumulative sales per year
cum_sale=np.cumsum(sales_data[:,1:],axis=1)
print("Cumulative sales per year",cum_sale)
```

```python
plt.figure(figsize=(6,6))
plt.plot(np.mean(cum_sale,axis=0))
plt.title("Average Cumulative Sales")
plt.xlabel("Sales")
plt.xlabel("Years")
plt.grid(True)
plt.show()
```

```python
# About Vectors
vector1=np.array([1,2,3,4,5])
vector2=np.array([6,7,8,9,10])

print("Vector Addition\t",vector1 + vector2)

print("Vector Multiplication\t",vector1 * vector2)

print("Dot Product\t",np.dot(vector1,vector2))

angle=np.arccos(np.dot(vector1,vector2)/(np.linalg.norm(vector1)*np.linalg.norm(vector2)))
print("Angle between them\t",angle)
```

```python
# vectorized operation means doing for all in a group
company_types=np.array(['KB Biryani','Bawarchi','RoyalHut','ChineaseHut'])
vectorized_upper=np.vectorize(str.upper)
print("Vectorized Upper\n",vectorized_upper(company_types))
```

```python
# Broadcast: here no need to use vectorize becaz numpy automatically knows to divide each year sale and each company sales by 12
monthly_avg=sales_data[:,1:] /12

print("Monthly Average\n",monthly_avg)
```
